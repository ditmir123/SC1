```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AQI Data set.csv', parse_dates=['Mounths'])
df.head()
```

```python
df.columns
```

```python
column_names = ['Id','Months','PM10','SO2','NOx','PM25','NH3','O3','CO','Benzene','AQI']
df.columns =column_names
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Months'].values, df['AQI'].values)
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('Air Quality Index (AQT) Trend over Time')
plt.xticks(rotation=75)
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Months'].values, df['PM10'].values, color='red')
plt.xlabel('Date')
plt.ylabel('PM10')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Months'].values,df['SO2'].values, color='purple')
plt.xlabel('Date')
plt.ylabel('SO2')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Months'].values,df['NOx'].values, color='blue')
plt.xlabel('Date')
plt.ylabel('NOx')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Months'].values,df['PM25'].values,color='yellow')
plt.xlabel('Date')
plt.ylabel('PM2.5')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Months'].values,df['CO'].values,color='violet')
plt.xlabel('Date')
plt.ylabel('CO')
plt.xticks(rotation=75)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(df['Months'].values,df['AQI'].values,color='skyblue',label='AQI')
plt.title('AQI Comparison across Months')
plt.xlabel('Months')
plt.ylabel('AQI Values')
plt.xticks(rotation=75)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
df.boxplot(color='blue')
plt.title('Distribution of AQI values for Pollutants')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df['AQI'],df['PM10'],c='red',label='PM10')
plt.scatter(df['AQI'],df['SO2'],c='blue',label='SO2')
plt.scatter(df['AQI'],df['NOx'],c='green',label='NOx')
plt.scatter(df['AQI'],df['PM25'],c='yellow',label='PM2.5')
plt.scatter(df['AQI'],df['CO'],c='purple',label='CO')
plt.scatter(df['AQI'],df['O3'],c='black',label='O3')
plt.title('Scatter plot: AQI levels vs Pollutants')
plt.xlabel('AQI')
plt.ylabel('Pollutant')
plt.legend()
plt.grid(True)
plt.show()
```
